```python
import requests 
from bs4 import BeautifulSoup
```

```python
#Making money green: Australia takes its first steps towards a net zero finance strategy
url = "https://theconversation.com/making-money-green-australia-takes-its-first-steps-towards-a-net-zero-finance-strategy-214063"
response = requests.get(url)
```

```python
soup=BeautifulSoup(response.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
paragraphs = soup.find_all('p')
```

```python
for p in paragraphs:
    if any(country in p.get_text() for country in ['Japan', 'Australia', 'China']):
        print(p)
```
