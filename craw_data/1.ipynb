```python
import requests 
from bs4 import BeautifulSoup
```

```python
#The New Australia–Japan Energy Relationship
url = "https://www.rieti.go.jp/users/shiro-armstrong/serial/007.html"
response = requests.get(url)
```

```python
soup=BeautifulSoup(response.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
paragraphs = soup.find_all('p')
```

```python
for p in paragraphs:
    if any(country in p.get_text() for country in ['Japan', 'Australia', 'China']):
        print(p)
```
