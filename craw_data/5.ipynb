```python
import requests 
from bs4 import BeautifulSoup
```

```python
#Australia has raised its climate targets and now needs to accelerate its clean energy transition, says new IEA review
url = "https://www.iea.org/news/australia-has-raised-its-climate-targets-and-now-needs-to-accelerate-its-clean-energy-transition-says-new-iea-review"
response = requests.get(url)
```

```python
soup=BeautifulSoup(response.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
paragraphs = soup.find_all('p')
```

```python
for p in paragraphs:
    if any(country in p.get_text() for country in ['Japan', 'Australia', 'China']):
        print(p)
```
