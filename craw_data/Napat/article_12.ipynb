```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
import spacy
from spacy import displacy
```

```python
url = "https://www.mri.co.jp/en/knowledge/insight/20240214.html"
response = requests.get(url)
```

```python
soup=BeautifulSoup(response.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
main_content = soup.find('div', {'id': 'contents_inner'})
for element in main_content(['a', 'em', 'span']):
    element.decompose()
```

```python
text_list = []
for paragraph in main_content.find_all('div', {'class': 'txt'}):
    paragraph_text = paragraph.get_text(separator=' ', strip=True)
    text_list.append(paragraph_text)
```

```python
df = pd.DataFrame(text_list, columns=["Paragraph Text"])
df.to_csv('12_paragraphs.csv', index=False)
```
