```python
import requests 
from bs4 import BeautifulSoup
```

```python
#Renewable Energy Investment in Australia
url = "https://www.rba.gov.au/publications/bulletin/2020/mar/renewable-energy-investment-in-australia.html"
response = requests.get(url)
```

```python
soup=BeautifulSoup(response.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
paragraphs = soup.find_all('p')
```

```python
for p in paragraphs:
    if any(country in p.get_text() for country in ['Japan', 'Australia', 'China']):
        print(p)
```
