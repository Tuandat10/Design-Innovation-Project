```python
import requests 
from bs4 import BeautifulSoup
```

```python
#China Invests Billions in International Renewable Energy Projects
url = "https://www.wri.org/insights/china-invests-billions-international-renewable-energy-projects"
response = requests.get(url)
```

```python
soup=BeautifulSoup(response.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
paragraphs = soup.find_all('p')
```

```python
for p in paragraphs:
    if any(country in p.get_text() for country in ['Japan', 'Australia', 'China']):
        print(p)
```
