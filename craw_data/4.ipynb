```python
import requests 
from bs4 import BeautifulSoup
```

```python
#Australian Budget commits A$25bn to clean energy and renewables projects
url = "https://international.austrade.gov.au/en/news-and-analysis/news/australian-budget-commits-a-25bn-to-clean-energy-and-renewables-"
response = requests.get(url)
```

```python
soup=BeautifulSoup(response.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
paragraphs = soup.find_all('p')
```

```python
for p in paragraphs:
    if any(country in p.get_text() for country in ['Japan', 'Australia', 'China']):
        print(p)
```
